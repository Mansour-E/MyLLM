```python
# get raw text from internet 
book  = requests.get('https://www.gutenberg.org/files/35/35-0.txt')

#extract just the text and habe a look at it 
text = book.text
print(type(text))
print(len(text))
```

```python
# character strings to replace with space
strings2replace = [
    '\r\n\r\nâ\x80\x9c', # new paragraph
    'â\x80\x9c', # open quote
    'â\x80\x9d', # close quote
    '\r\n', # new line
    'â\x80\x94', #hyphen
    'â\x80\x99', # single apostrophe
    'â\x80\x98', # single quote
    '_',# underscore , used for stressing
]
# use regular expression (re) to repace those strings with space 
for str2match in strings2replace:
    regexp = re.compile(r'%s'%str2match)
    text = regexp.sub(' ', text)

# remove non-ASCII characters
text = re.sub(r'[^\x00-\x7f]+', ' ', text)

# removing numbers
text = re.sub(r'\d+', '' , text)

# and make everything lower-case
text = text.lower()

# let's have a look!
text[:2000]
```

```python
# split by punctuation
import string
print(string.punctuation)
puncts4re = rf'[{string.punctuation}\s]+'

words = re.split(puncts4re, text)
words = [item.strip() for item in words if item.strip()]

# remove single-character words 
words = [item for item in words if len(item)>1]

# let's hve a look
words[:50]
```

```python
# create the vocab! ( set of unique words)
vocab = sorted(set(words))

# convenience variable for later
nWords = len(words)
nLex = len(vocab)

print(f'{nWords} words')
print(f'{nLex} unique tokens')
```

<h4>Create token dictionaries and encodeer/decoder functions</h4>

```python
word2idx = {w:i for i , w in enumerate(vocab)}
idx2word = {i:w for i , w in enumerate(vocab)}

# print out a few 
for i in list(word2idx.items())[0:10000:87]:
    print(i)
```

```python
np.zeros(4,dtype=int)
```

```python
# encoder function (using for-loop instead of list-comp)
def encoder(words, encode_dict):
    
    #initialize a vector of numerical indices numpy array
    idxs = np.zeros(len(words), dtype=int)
    
    # loop through the words and find their token in the vocab
    for i , word in enumerate(words):
        idxs[i] = encode_dict[word]
        
    #return the indices    
    return idxs

# also need a decoder function
def decoder(idxs, decode_dict):
    
    return ' '.join([decode_dict[i] for i in idxs])
```

```python
# test the encoder 
print(encoder(['the','time','machine'] , word2idx))

# test the decoder 
print(decoder([1,2,3], idx2word))
```

```python
# random start location
startidx = np.random.choice(nWords)

# sequential word indices
idxs = np.arange(startidx, startidx+10)

print('Word indices: ')
print(idxs) , print('')

print('the words : ')
wordseq = [words[i] for i in idxs]
print(wordseq) , print('')

print('Token indices:')
tokenseq = encoder(wordseq ,word2idx)
print(tokenseq) , print('')

# decode 
print('decode text from indices:')
decoder(tokenseq, idx2word)
```

```python
import numpy as np 
import matplotlib.pyplot as plt

import requests
import re
import string

import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
text = requests.get('https://www.gutenberg.org/files/35/35-0.txt').text

strings2replace = [
    '\r\n\r\nâ\x80\x9c', # new paragraph
    'â\x80\x9c', # open quote
    'â\x80\x9d', # close quote
    '\r\n', # new line
    'â\x80\x94', #hyphen
    'â\x80\x99', # single apostrophe
    'â\x80\x98', # single quote
    '_',# underscore , used for stressing
]

for i in strings2replace:
    text = re.compile(r'%s'%i).sub(' ',text)
    
text = re.sub(r'[^\x00-\x7f]+', ' ', text)
text = re.sub(r'\d+', '' , text).lower()
```

```python
text
```

```python
# split into words that contain > 1 character 
words = re.split(fr'[{string.punctuation}\s]+', text)
words = [item.strip() for item in words if item.strip()]
words = [item for item in words if len(item) > 1]

vocab = sorted(set(words))
nWords = len(words)
nLex = len(vocab)
```

```python
word2idx = { w:i for i , w in enumerate(vocab)}
idx2word = {i:w for i , w in enumerate(vocab)}

```

```python
# create encoder and decoder function
def encoder(words, encode_dict):
    
    idxs = np.zeros(len(words), dtype=int)
    for i , word in enumerate(words):
        idxs[i] = encode_dict[word]
    return idxs

def decoder(idxs , decode_dict):
    return ' '.join([decode_dict[i] for i in idxs])
```

```python
randomTokens = np.random.randint(0 , len(vocab) , 10)
print(f'Random tokens: \n\t{randomTokens}\n')
print(f'decoder text: \n\t{decoder(randomTokens, idx2word)}')
```

```python
brownNoise = np.cumsum(np.random.choice([-1,1], 3000))
plt.figure(figsize=(10,3))
plt.plot(brownNoise,'k')
plt.gca().set(xlim=[0,len(brownNoise)],xlabel='"Time" (?)', ylabel='Signal amplitude')
plt.show()
```

```python
brownNoise = np.cumsum(np.random.choice([-1,1], 30))
print(brownNoise)

BrownianRandomTokens = brownNoise + np.random.choice(nLex , 1)
print(BrownianRandomTokens)

print(' ')

print(f'Bownian random tokens: \n\t{BrownianRandomTokens}\n')
print(f'Decoded text: \n\t{decoder(BrownianRandomTokens, idx2word)}')
    
```
