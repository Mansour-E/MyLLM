```python
text = ['All that we are is the result of what we have thought',
        'To be or not to be that is the question',
        'Be yourself everyone else is already taken']
```

```python
words = set((' '.join(text).lower()).split())
```

```python
word2idx = {}
for i , word in enumerate(words):
        word2idx[word] = i
word2idx      
```

```python
idx2word = {}
for i , word in enumerate(words):
        idx2word[i] = word
idx2word
```
